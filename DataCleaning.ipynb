```python
#using IQR to handle outliers
for item in df.columns:
    Q1 = df[item].quantile(0.25)
    Q3 = df[item].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    df = df[(df[item] >= lower_bound) & (df[item] <= upper_bound)]
```
