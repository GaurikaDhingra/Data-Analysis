```python
#Create a Pairplot to understand relation between variables
sns.pairplot(df, hue='Species') #pairplot of data will be made with hue in respect to species column
plt.suptitle("Pairplot for Iris Dataset")
plt.show()
```

KEY FINDINGS

Comparing the lengths and widths of each species, we canfind that Versicolor and Virginica are found to be close to each other.

We can observe that Sepal Length and Sepal Width are more closely related as compared to relationship between Petal Length and Petal Width


By Using PairPlots we are able to visualize the relationship of different features among each other in the Iris Dataset.

```python
#Correlation Heatmap
# Select numeric columns only
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Compute correlation matrix
corr_matrix = numeric_df.corr()

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

The correlation matrix is used to understand the correlation amobng various fetures.

We can see a negative correlation with Sepal Width and Sepal Length which means if one increases the ther one reduces. A negative correlation can also be seen among Sepal Width and Petal Length and sepal width along with petal width.
Thus, we can say that Sepal Width tends to have a negative correlation with all other features of the data.

Other than this, we can see a positive correlation between all other features indicating a positive change in one factor due to another factor.

```python
# Group by 'Species' and sum the 'SepalLength' for each species, then sort the values in descending order
grp1 = df.groupby('Species', as_index=False)['SepalLengthCm'].sum().sort_values(by='SepalLengthCm', ascending=False)

# Plot BarPlot
plt.plot(figsize = (3,5))
plt.bar(grp1['Species'], grp1['SepalLengthCm'], width = 0.5, color=['r', 'g', 'b'])
plt.title("Sepal Length Comparison")
plt.xlabel("Species")
plt.ylabel("Total Sepal Length")
plt.show()
```

```python
# Group by 'Species' and sum the 'SepalWidth' for each species, then sort the values in descending order
grp2 = df.groupby('Species', as_index=False)['SepalWidthCm'].sum().sort_values(by='SepalWidthCm', ascending=False)

# Plot BarPlot
plt.plot(figsize = (3,5))
plt.bar(grp2['Species'], grp2['SepalWidthCm'], width = 0.5, color=['r', 'g', 'b'])
plt.title("Sepal Width Comparison")
plt.xlabel("Species")
plt.ylabel("Total Sepal Width")
plt.show()
```

```python
# Group by 'Species' and sum the 'PetalLength' for each species, then sort the values in descending order
grp3 = df.groupby('Species', as_index=False)['PetalLengthCm'].sum().sort_values(by='PetalLengthCm', ascending=False)

# Plot BarPlot
plt.plot(figsize = (3,5))
plt.bar(grp3['Species'], grp3['PetalLengthCm'], width = 0.5, color=['r', 'g', 'b'])
plt.title("Petal Length Comparison")
plt.xlabel("Species")
plt.ylabel("Total Petal Length")
plt.show()
```

```python
# Group by 'Species' and sum the 'PetalWidth' for each species, then sort the values in descending order
grp4 = df.groupby('Species', as_index=False)['PetalWidthCm'].sum().sort_values(by='PetalWidthCm', ascending=False)

# Plot BarPlot
plt.plot(figsize = (3,5))
plt.bar(grp4['Species'], grp4['PetalWidthCm'], width = 0.5, color=['r', 'g', 'b'])
plt.title("Petal Width Comparison")
plt.xlabel("Species")
plt.ylabel("Total Petal Width")
plt.show()
```

```python
#Combining the comparison graphs for better analysis
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
plt.suptitle("Comapritive Analysis of Iris Dataset")

#plot 1
# Group by 'Species' and sum the 'SepalLength' for each species, then sort the values in descending order
grp1 = df.groupby('Species', as_index=False)['SepalLengthCm'].sum().sort_values(by='SepalLengthCm', ascending=False)

# Plot BarPlot
plt.subplot(2,2,1)
plt.bar(grp1['Species'], grp1['SepalLengthCm'], width = 0.5, color=['r', 'g', 'b'])
plt.title("Sepal Length Comparison")
plt.xlabel("Species")
plt.ylabel("Total Sepal Length")

#Plot 2
# Group by 'Species' and sum the 'SepalWidth' for each species, then sort the values in descending order
grp2 = df.groupby('Species', as_index=False)['SepalWidthCm'].sum().sort_values(by='SepalWidthCm', ascending=False)

# Plot BarPlot
plt.subplot(2,2,2)
plt.bar(grp2['Species'], grp2['SepalWidthCm'], width = 0.5, color=['r', 'g', 'b'])
plt.title("Sepal Width Comparison")
plt.xlabel("Species")
plt.ylabel("Total Sepal Width")

#Plot 3
# Group by 'Species' and sum the 'PetalLength' for each species, then sort the values in descending order
grp3 = df.groupby('Species', as_index=False)['PetalLengthCm'].sum().sort_values(by='PetalLengthCm', ascending=False)

# Plot BarPlot
plt.subplot(2,2,3)
plt.bar(grp3['Species'], grp3['PetalLengthCm'], width = 0.5, color=['r', 'g', 'b'])
plt.title("Petal Length Comparison")
plt.xlabel("Species")
plt.ylabel("Total Petal Length")

#Plot 4
# Group by 'Species' and sum the 'PetalWidth' for each species, then sort the values in descending order
grp4 = df.groupby('Species', as_index=False)['PetalWidthCm'].sum().sort_values(by='PetalWidthCm', ascending=False)

# Plot BarPlot
plt.subplot(2,2,4)
plt.bar(grp4['Species'], grp4['PetalWidthCm'], width = 0.5, color=['r', 'g', 'b'])
plt.title("Petal Width Comparison")
plt.xlabel("Species")
plt.ylabel("Total Petal Width")

#Add layout
plt.tight_layout(pad=3.0)  # Increase padding between subplots

#Display Graph
plt.show()
```

KEY FINDINGS

To visualise the comparison of Lengths and widths of the three species, we used groupby function and used the sum of the lengths for our analysis.

We can visualize that the total Sepal Length of the three species varies between 250cm to 300cm, and the total Sepal Length of the Virginica Species is the highest at 300cm and Setosa's is the lowest at the value of 250 cm.

A similar ratio of change can be seen in Sepal Width of the three species. The total sepal width varies from 150 cm to 130 cm. Setosa's total is the highest of about 150 cm whereas the lowest sepal width as an aggregate can be seen of versicolor of around 130 cm.

Petal Length and Petal Width whereas can see a change in the comparison.
The highest of the total Petal Length is about 250 cm which is of Virginica Species and the lowest total saw a major drop to about 80 cm which is of Setosa Species.

A similar ratio can be seen for Petal Width comparison as well, the highest of the total Petal Width comes out to be about 100 cm of Virginica Species and similar to the Petal Length Analysis, the lowest of the total petal Width also comes out to be of Setosa Species of only about 10cm.


```python
#CREATING DASHBOARD

# Create a new figure with a larger size
fig = plt.figure(figsize=(20, 15))
# Define subplot grid with reduced sizes and padding
grid = fig.add_gridspec(2, 4, hspace=0.5, wspace=0.5)

# Plot 1: Countplot of Species
ax1 = fig.add_subplot(grid[0, 0])
sns.countplot(x='Species', data=df, ax=ax1)
ax1.set_title("Iris-Distribution")
ax1.set_xlabel("Species")
ax1.set_ylabel("Count")

# Plot 2: Pie chart of Species distribution
ax2 = fig.add_subplot(grid[0, 1])
species_counts = df["Species"].value_counts()
ax2.pie(species_counts, labels=species_counts.index, autopct="%.1f%%", startangle=140)
ax2.set_title("Flower Species Distribution")
ax2.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Plot 3: Correlation Heatmap
ax3 = fig.add_subplot(grid[0, 2:])
numeric_df = df.select_dtypes(include=['float64', 'int64'])
corr_matrix = numeric_df.corr()
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", ax=ax3)
ax3.set_title("Correlation Heatmap")

# Plot 4: Comparative Analysis
# Plot 1: Sepal Length Comparison
ax4 = fig.add_subplot(grid[1, 0])
ax4.bar(grp1['Species'], grp1['SepalLengthCm'], width=0.5, color=['r', 'g', 'b'])
ax4.set_title("Sepal Length Comparison")
ax4.set_xlabel("Species")
ax4.set_ylabel("Total Sepal Length")

# Plot 2: Sepal Width Comparison
ax5 = fig.add_subplot(grid[1, 1])
ax5.bar(grp2['Species'], grp2['SepalWidthCm'], width=0.5, color=['r', 'g', 'b'])
ax5.set_title("Sepal Width Comparison")
ax5.set_xlabel("Species")
ax5.set_ylabel("Total Sepal Width")

# Plot 3: Petal Length Comparison
ax6 = fig.add_subplot(grid[1, 2])
ax6.bar(grp3['Species'], grp3['PetalLengthCm'], width=0.5, color=['r', 'g', 'b'])
ax6.set_title("Petal Length Comparison")
ax6.set_xlabel("Species")
ax6.set_ylabel("Total Petal Length")

# Plot 4: Petal Width Comparison
ax7 = fig.add_subplot(grid[1, 3])
ax7.bar(grp4['Species'], grp4['PetalWidthCm'], width=0.5, color=['r', 'g', 'b'])
ax7.set_title("Petal Width Comparison")
ax7.set_xlabel("Species")
ax7.set_ylabel("Total Petal Width")


# # Adjust layout
# plt.tight_layout()

# Show the dashboard
plt.show()
```
